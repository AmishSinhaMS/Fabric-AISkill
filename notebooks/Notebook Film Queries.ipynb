# Notebook Film Queries
Run some queries against the two film data tables

## Queries

### Count records
Should be 250 for both tables

```python
%%sql
SELECT 'films250' s, COUNT(*) c FROM films250
UNION ALL
SELECT 'filmsgross', COUNT(*) FROM filmsgross;
```

### Find highly rated films not in the top 250 Gross

```python
%%sql
SELECT primaryTitle
FROM films250 f
EXCEPT
SELECT primaryTitle
FROM filmsgross
ORDER BY 1;
```

### Find high grossing films not in the top 250 rated

```python
%%sql
SELECT primaryTitle
FROM filmsgross
EXCEPT
SELECT primaryTitle
FROM films250 f
ORDER BY 1;
```

```python
%%sql
SELECT *
FROM filmsgross
WHERE primaryTitle = 'Barbie';
```

```python
%%sql
SELECT *
FROM films250
WHERE primaryTitle = 'Barbie'
```

```python
%%sql
SELECT *
FROM films250
WHERE primaryTitle Like '%Avatar%';
```

```python
%%sql
SELECT *
FROM filmsgross
WHERE primaryTitle Like '%Avatar%'
```
